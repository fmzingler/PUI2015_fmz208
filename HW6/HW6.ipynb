#Task 1:
#Event detection: Identify the most prominent event. There is a very significant drop (>3-sigma) in all time series.
#Identify it and figure out what it is due to.

##Select fares: all the fares for 30-days passes (assuming most common type of fare)

```python
fares_all = np.nansum(fares, axis=0)
fares_30D = fares_all[6]
```

##Calculate mean and std
#plot with the threshold

```python
mean30D = np.nanmean(fares_30D)
std30D = np.nanstd(fares_30D)
thresholds = (mean30D - std30D*3, mean30D + std30D*3)

weeks = pd.date_range(start = '2010-05-21', periods = 194, freq = 'W')

fig = pl.figure(figsize=(12, 12))
pl.plot(weeks, fares_30D)

pl.plot([weeks.min(), weeks.max()], [
        thresholds[1], thresholds[1]], 'red', label='Max Threshold (+3sig)')

pl.plot([weeks.min(), weeks.max()], [
        thresholds[0], thresholds[0]], 'orange', label='Min Threshold (-3sig)')

pl.title("Trips for the 30-days passes")
pl.ylabel("Number of Trips")
pl.legend()
```

As we can see, a few of the events are outside the minimum threshold, indicating that some events are causing a drop on the trips. THe most important one would be between September 2012 and January 2013, which includes the period when  Hurricane Sandy hits the city (October 2012). 
Other events are occurring before January every year where a significant drop on the trips occurs. Since the data corresponds to 30-days passes, usually purchased by commuters and people who uses everyday the subway, this could be related to holidays (Christmas), since it's occurring every yeaar even if it's not below the minimum threshold.
